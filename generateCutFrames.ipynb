## Extract Images froom videos and move to folders

```python
#  References:
# https://www.kaggle.com/code/willstone98/transfer-learning-fine-tuning-with-vgg16

# Imports
from keras.preprocessing.image import load_img
# convert the image pixels to a numpy array
from keras.preprocessing.image import img_to_array
# Imports
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
# from tensorflow.python.keras import ImageDataGenerator
```

```python

#  Constants
SAVING_FRAMES_PER_SECOND = 5

PATH_VID_TRAIN = '../videos/train' # Just to didn't test with many videos
PATH_VID_TEST = '../videos/test'
PATH_FRAMES_TRAIN = 'frames/train'
PATH_FRAMES_TEST = 'frames/test'

# Load data
# !dir
dataset_path = os.listdir(PATH_VID_TRAIN)
label_types = os.listdir(PATH_VID_TRAIN) # if we use the folders to separate what is
print (label_types)  

boxCh12 = (100, 830, 1018, 1280)
```

```python
#  methods
def extract_images(video_path, frames_folder):
    vidcap = cv2.VideoCapture(video_path)
    success,image = vidcap.read()
    count = 0
    # video_path.split('/')[-1][:-4] get file name
    while success:
        x, y, w, h = boxCh12
        image = image[x:y, w:h]
        cv2.imwrite(frames_folder + "\\" + video_path.split('/')[-2]  + "\\"+ video_path.split('/')[-1][:-4]+ "_%d.jpg" % count, image)     # save frame as JPEG file      
        success,image = vidcap.read()
        count += 1

def preProcess(img_list, path):
    for item in os.listdir(path):
        # print(path + '/' + item )
        img = load_img(path +'/' + item ,target_size=(224, 224))
        img = img_to_array(img)
        # reshape data for the model
        img = img.reshape((img.shape[0], img.shape[1], img.shape[2]))
        # img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))
        # img = preprocess_input(img)
        # predict the probability across all output classes
        img_list.append(img)

def preProcessByImage(img_list, frame):
        img = load_img(frame ,target_size=(224, 224))
        img = img_to_array(img)
        # reshape data for the model
        img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))
        # img = preprocess_input(img)
        # predict the probability across all output classes
        img_list.append(img)
```

```python
# Preparing Training Data
for item in dataset_path:
    # Get all the file names
    all_rooms = os.listdir(PATH_VID_TRAIN)
# Add them to the path list
rooms = []
for room in all_rooms:
    all_files_from_class = os.listdir(PATH_VID_TRAIN+ '/' + room)
    for file in all_files_from_class:
        rooms.append(str(PATH_VID_TRAIN) + '/' + room + '/' + file)
print('All rooms: ', rooms)
for room in rooms:
    extract_images(room, PATH_FRAMES_TRAIN)

# Preparing Testing Data
for item in dataset_path:
    # Get all the file names
    all_rooms = os.listdir(PATH_VID_TEST)
# Add them to the path list
rooms = []
for room in all_rooms:
    all_files_from_class = os.listdir(PATH_VID_TEST+ '/' + room)
    for file in all_files_from_class:
        rooms.append(str(PATH_VID_TEST) + '/' + room + '/' + file)
print('All rooms: ', rooms)
for room in rooms:
    extract_images(room, PATH_FRAMES_TEST)
```

```python
#  PRE PROCESSING
images_train = []

dataset_path = os.listdir(PATH_FRAMES_TRAIN)
# Preparing Training Data
for item in dataset_path:
    preProcess(images_train,PATH_FRAMES_TRAIN + '/' + item)

images_test = []
dataset_path = os.listdir(PATH_FRAMES_TEST)
for item in dataset_path:
    preProcess(images_test,PATH_FRAMES_TEST + '/' + item)

images_train
```
